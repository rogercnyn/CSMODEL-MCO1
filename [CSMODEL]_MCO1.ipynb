### **1. How many YouTube channels are there per country?**

**Objective:**

The first exploratory question aims to examine the number of YouTube channels per country. Understanding the distribution of YouTube channels across different countries can provide insights into the platform's global reach and popularity. It also helps identify key markets for content creators and marketers.

**Methodology:**

The dataset contains country names stored as strings in the Country column. To visualize the number of YouTube channels per country, we extracted unique countries with relevant attributes and dropped duplicates. We then counted the number of channels for each country, merged the two dataframes to include the channel count, and sorted the results by Channel_Count. A bar plot was employed to visualize the data.

```python
# Extracting unique countries with relevant attributes and dropping duplicates
unique_countries = youtube_df[['Country', 'Abbreviation']].drop_duplicates(subset=['Country', 'Abbreviation'])

# Counting the number of channels for each country
country_channel_counts = youtube_df.groupby(['Country', 'Abbreviation']).size().reset_index(name='Channel_Count')

# Merging the two dataframes to include the channel count
unique_countries = pd.merge(unique_countries, country_channel_counts, on=['Country', 'Abbreviation'], how='left')

# Sorting unique_countries by Channel_Count
unique_countries = unique_countries.sort_values(by='Channel_Count', ascending=False)

# Plotting the data
plt.figure(figsize=(14, 12))
sns.barplot(x='Channel_Count', y='Country', data=unique_countries, palette='viridis', hue='Country')
plt.title('Number of YouTube Channels per Country')
plt.xlabel('Number of YouTube Channels')
plt.ylabel('Country')
plt.xticks(rotation=45)
plt.show()

# Measures of Central Tendency and Dispersion
channel_counts = unique_countries['Channel_Count']

mean = channel_counts.mean()
median = channel_counts.median()
mode = channel_counts.mode()[0]  # mode() returns a Series
range_ = channel_counts.max() - channel_counts.min()
variance = channel_counts.var()
std_dev = channel_counts.std()

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Range: {range_}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
```

#### *Conclusion for Exploratory Question 1*

In conclusion, the analysis of the number of YouTube channels per country reveals several key insights. The United States leads with the highest number of YouTube channels, followed closely by India and Brazil. This trend is reflective of the substantial user bases and high internet penetration rates in these countries, making them prime markets for YouTube content. The bar plot visualization and numerical summary (mean, median, mode, range, variance, and standard deviation) of channel counts across different countries suggest diverse levels of engagement with the platform. For content creators and marketers, these findings highlight important geographic areas to focus on for reaching larger audiences. Additionally, understanding which countries have fewer channels can present opportunities for new creators to enter less saturated markets.

High unemployment rates can influence the amount of free time individuals have, potentially leading to higher content consumption or creation as a means of income.

### **3. What channel has the highest views for each category?**


**Objective:**

The third exploratory question aims to identify which channels have the highest views in each category. This information is valuable for understanding the leading content creators in different niches and the types of content that attract the most views. By recognizing the top-performing channels, new content creators can identify trends and strategies that contribute to high view counts.

**Methodology:**

The dataset contains category names and video view counts. To determine the channel with the highest views in each category, we grouped the data by category, found the channel with the highest views within each group, and formatted the view counts for better readability. The results were displayed in a formatted table.

```python
def format_with_commas(num):
    # Convert scientific notation to float and then format with commas
    return "{:,.0f}".format(float(num))

# Group by category and find channel with highest views
highest_views_per_category = youtube_df.groupby('category').apply(lambda x: x.loc[x['video views'].idxmax()])

video_views = highest_views_per_category['video views']

# Format 'video views' column with commas
highest_views_per_category['video views'] = highest_views_per_category['video views'].apply(format_with_commas)

highest_views_table = highest_views_per_category[['Youtuber', 'video views']]

table = tabulate(highest_views_table, headers='keys', tablefmt='fancy_grid')

print(table)

mean_views = video_views.mean()
median_views = video_views.median()
mode_views = video_views.mode()[0]  # mode() returns a Series
range_views = video_views.max() - video_views.min()
variance_views = video_views.var()
std_dev_views = video_views.std()

print(f"Mean: {mean_views}")
print(f"Median: {median_views}")
print(f"Mode: {mode_views}")
print(f"Range: {range_views}")
print(f"Variance: {variance_views}")
print(f"Standard Deviation: {std_dev_views}")
```

#### *Conclusion for Exploratory Question 3*

In conclusion, the analysis of the channels with the highest views in each category provides a clear view of the leading performers on YouTube. The formatted table visualization and numerical summary (mean, median, mode, range, variance, and standard deviation) of video views offer benchmarks and targets for aspiring content creators and marketers. These top channels represent the most successful content creators in their respective categories, attracting the highest view counts. By studying these top channels, new creators can gain insights into effective content strategies, popular topics, and audience engagement techniques.



### **4. How do the total subscribers vary among YouTube channels created in different years?**


**Objective:**

The fourth exploratory question aims to understand how the total number of subscribers varies among YouTube channels created in different years.


**Methodology:**

The dataset contains the year a channel was created and the number of subscribers. To analyze this, the data was grouped by the year the channel was created, and the average number of subscribers for each year was calculated. Following this, the variation in average subscribers for channels created in different years was plotted. Measures of central tendency (mean, median, mode) and measures of dispersion (range, variance, standard deviation, interquartile range) for the subscriber counts were also calculated. Finally, the Pearson correlation coefficient was determined to understand the strength and direction of the relationship between the year of creation and the number of subscribers.

```python
yearly_subs = youtube_df.groupby('created_year')['subscribers'].mean().reset_index()

# Plot the variation of subscribers for channels created in different years
plt.figure(figsize=(12, 8))
sns.lineplot(x='created_year', y='subscribers', data=yearly_subs)
plt.title('Variation of Subscribers for YouTube Channels Created in Different Years')
plt.xlabel('Year Created')
plt.ylabel('Average Subscribers')
plt.xticks(yearly_subs['created_year'].astype(int))
plt.show()

# Measures of Central Tendency
mean_subscribers = yearly_subs['subscribers'].mean()
median_subscribers = yearly_subs['subscribers'].median()
mode_subscribers = yearly_subs['subscribers'].mode()[0]  # Mode can return multiple values, take the first one

print(f"Mean Subscribers: {mean_subscribers}")
print(f"Median Subscribers: {median_subscribers}")
print(f"Mode Subscribers: {mode_subscribers}")

# Measures of Dispersion
range_subscribers = yearly_subs['subscribers'].max() - yearly_subs['subscribers'].min()
variance_subscribers = yearly_subs['subscribers'].var()
std_deviation_subscribers = yearly_subs['subscribers'].std()
iqr_subscribers = yearly_subs['subscribers'].quantile(0.75) - yearly_subs['subscribers'].quantile(0.25)

print(f"Range of Subscribers: {range_subscribers}")
print(f"Variance of Subscribers: {variance_subscribers}")
print(f"Standard Deviation of Subscribers: {std_deviation_subscribers}")

# Correlation between created_year and subscribers
correlation = yearly_subs['created_year'].corr(yearly_subs['subscribers'])

print(f"Pearson Correlation between Created Year and Subscribers: {correlation}")
```

#### *Conclusion for Exploratory Question 4*

In conclusion, the analysis of subscriber variation among YouTube channels created in different years reveals a clear trend. Channels created in earlier years generally have higher subscriber counts, as indicated by the declining trend in the average number of subscribers over time. The line plot visualization and numerical summary (mean, median, mode, range, variance, standard deviation, and interquartile range) provide a comprehensive understanding of this variation. The strong negative correlation between the year of creation and total subscribers suggests that earlier channels benefit from significantly higher subscriber counts. This trend can be attributed to factors such as having more time to grow, facing less competition, and potentially benefiting from YouTube's earlier content discovery algorithms. Overall, the data suggests that being an early adopter on YouTube has provided a substantial advantage in accumulating subscribers.

### **6. What is the relationship between the number of subscribers and total video views?**

**Objective:**

The sixth exploratory question aims to identify the relationship between the number of subscribers and total video views. Understanding this relationship can help determine the impact of subscriber count on overall viewership and guide content creators in their growth strategies.

**Methodology:**

The dataset contains the number of subscribers and total video views for each channel. To analyze the relationship, we employed a scatter plot to visualize the correlation between these variables. We also calculated the Pearson correlation coefficient to quantify the strength of the relationship.

```python
# Scatter plot to visualize the relationship between subscribers and video views
plt.figure(figsize=(8, 6))
sns.scatterplot(x='subscribers', y='video views', data=youtube_df, alpha=0.7)
plt.title('Relationship between Subscribers and Video Views')
plt.xlabel('Number of Subscribers')
plt.ylabel('Total Video Views')

# Use ticklabel_format to prevent scientific notation
plt.ticklabel_format(style='plain', axis='both', useOffset=False)

# Format the tick labels to include commas
formatter = ticker.FuncFormatter(lambda x, p: format(int(x), ','))
plt.gca().xaxis.set_major_formatter(formatter)
plt.gca().yaxis.set_major_formatter(formatter)
plt.grid(True)
plt.show()

# Correlation analysis
correlation = youtube_df['subscribers'].corr(youtube_df['video views'])
print(f"Pearson correlation coefficient: {correlation}")
```

#### *Conclusion for Exploratory Question 6*

In conclusion, the scatter plot reveals a **positive correlation** between the number of subscribers and total video views. As the number of subscribers increases, the total video views also tend to increase. A significant clustering of channels is evident in the lower-left corner of the plot, indicating that most channels have a moderate number of subscribers and video views, especially in comparison to the extreme outliers. There are a few **outliers** with extremely high numbers of subscribers and video views, standing out from the general cluster of data points. Additionally, there is noticeable **variability** in video views among channels with a similar number of subscribers. Some channels with fewer subscribers achieve disproportionately high video views, and vice versa.

The strong positive correlation **(0.825)** indicates that channels with higher subscriber counts generally have higher total video views. This relationship underscores the importance of growing a subscriber base to boost overall viewership. However, the variability and presence of outliers suggest that other factors, such as content quality, viewer engagement, and viral potential, also play significant roles in driving video views.

Outliers were not removed to maintain a comprehensive view of the dataset, including extreme cases that reflect real-world phenomena on YouTube. Removing outliers could distort the variability and exclude valuable insights into channel dynamics and audience engagement. Retaining outliers ensures transparency, robustness, and a balanced perspective in the data analysis.

**Methodology:**

The dataset contains several numerical columns of interest. To analyze the relationships between these variables, we selected the relevant columns and calculated the correlation matrix. We visualized the correlation matrix using a heatmap to highlight significant correlations.

```python
# Selecting relevant columns for the correlation matrix
relevant_columns = ['subscribers', 'video views', 'Urban_population', 'Unemployment rate', 'uploads']
relevant_df = youtube_df[relevant_columns]

# Calculating the correlation matrix
plt.figure(figsize=(12, 10))
corr_matrix = relevant_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix for Relevant Variables')
plt.show()
```

**Findings:**


1. Subscribers and Urban Population: A positive correlation suggests that channels from regions with higher urban population percentages tend to have more subscribers, aligning with the bar plot analysis.

2. Subscribers and Unemployment Rate: A negative correlation indicates that countries with higher unemployment rates tend to have fewer subscribers, supporting the findings from the bar plot analysis.

3. Subscribers and Uploads: A positive correlation suggests that channels with more uploads tend to have more subscribers, indicating that content quantity may contribute to subscriber growth.

4. Video Views and Urban Population: A positive correlation suggests that regions with higher urban population percentages tend to have channels with more video views.

5. Video Views and Unemployment Rate: A negative correlation indicates that countries with higher unemployment rates tend to have channels with fewer video views.

6. Uploads and Video Views: A positive correlation suggests that channels with more uploads tend to have higher total video views.

7. Subscribers and Video Views: A high positive correlation (close to 1) indicates that channels with more subscribers tend to have higher total video views, confirming the relationship observed in the scatter plot analysis.





#### *Conclusion for the correlation matrix*

In conclusion, the correlation matrix analysis of the Global YouTube Statistics 2023 dataset reveals the following insights:

1. Strong Positive Correlation Between Subscribers and Video Views: Channels with more subscribers tend to have significantly higher total video views. This relationship is consistent across various categories and indicates that subscriber count is a strong predictor of video view count.

2. Positive Impact of Urban Population on Subscribers and Video Views: Channels originating from regions with higher urban population percentages tend to have more subscribers and higher total video views. This suggests that urbanization positively influences the popularity and performance of YouTube channels.
